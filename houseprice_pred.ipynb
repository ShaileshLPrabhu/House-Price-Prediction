Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from  sklearn.model_selection import train_test_split
```

Importing the Dataset

```python
data= pd.read_csv("housing.csv")
```

Cleaning Data

```python
data.info()
```

```python
#deleting rows with null values
data.dropna(inplace= True)
```

```python
#deleting rows with duplicate  values in a dataframe
data.drop_duplicates(inplace=True)
```

```python
#data after cleaning
data.info()
```

spliting the data into training and testing

```python
X=data.drop(['median_house_value'],axis=1)
Y=data['median_house_value']

X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size= 0.2)
```

```python
train_data=X_train.join(Y_train)
```

```python
train_data.hist(figsize=(10,8))
```

Exploring data and preprocessing it

```python
plt.figure(figsize=(15,8))
sn.heatmap(train_data.corr(numeric_only=True), cmap="YlGnBu", fmt=".2f")
```

```python
train_data['total_rooms']= np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']= np.log(train_data['total_bedrooms']+1)
train_data['population']= np.log(train_data['population']+1)
train_data['households']= np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
plt.figure(figsize=(15, 8))
sn.heatmap(train_data.corr(), annot= True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15, 8))
sn.scatterplot(x="latitude",y="longitude",data=train_data,hue="median_house_value",palette="coolwarm")
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15, 8))
sn.heatmap(train_data.corr(), annot= True, cmap="YlGnBu")
```

Building a Linear Regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

X_train,Y_train= train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
X_train_s= scaler.fit_transform(X_train)

Lreg=LinearRegression()
Lreg.fit(X_train_s,Y_train)    
```

```python
test_data=X_test.join(Y_test)

test_data['total_rooms']= np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']= np.log(test_data['total_bedrooms']+1)
test_data['population']= np.log(test_data['population']+1)
test_data['households']= np.log(test_data['households']+1)

test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms']/test_data['households']

X_test,Y_test= test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```

```python
test_data
```

```python
X_test_s = scaler.transform(X_test)
```

```python
print("Linear Regression Train Score:", Lreg.score(X_train_s,Y_train))
print("Linear Regression Test Score:", Lreg.score(X_test_s,Y_test))
```

```python
from sklearn.ensemble  import RandomForestRegressor

r_frst=RandomForestRegressor(n_estimators = 10, random_state = 42)
#print("Training the model")

r_frst.fit(X_train, Y_train) # Training the algorithm on the training data
```

```python
print("Random Forest Train Score:", r_frst.score(X_train,Y_train))
print("Random Forest Test Score:", r_frst.score(X_test,Y_test))
```
